```python
import random
import numpy as np
import pandas as pd
import sqlite3
from IPython.display import display
```

```python
adult_data_df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data')
display(adult_data_df.head(5))
```

```python
adult_data_df.columns =["age","workclass","fnlwgt","education","education_num","marital_status","occupation","relationship","race","sex","capital_gain","capital_loss","hours_per_week","native_country","income"]
display(adult_data_df.head(5))
```

1.Create an sqlalchemy engine using a sample from the data set

```python
import sqlalchemy
from sqlalchemy import create_engine
engine = create_engine('sqlite:///sqladb', echo=False)
adult_data_df.to_sql('sqladb', engine, if_exists='replace')

connection  = sqlite3.connect("sqladb")
cursor = connection.cursor()
```

2.Write two basic update queries

```python
cursor.execute('UPDATE sqladb SET fnlwgt = "284788" WHERE fnlwgt = " 284582"')
print(pd.read_sql_query("SELECT * FROM sqladb", connection).head(5))
```

Second update:

```python
cursor.execute('UPDATE sqladb SET workclass = "Public Sector" WHERE workclass = " Private"')
print(pd.read_sql_query("SELECT * FROM sqladb", connection).tail(6))
```

3.Write two delete queries

```python
cursor.execute('DELETE FROM sqladb WHERE fnlwgt = " 234721"')
print(pd.read_sql_query("SELECT * FROM sqladb", connection).head(5))
```

Second deletion:

```python
cursor.execute('DELETE FROM sqladb WHERE age = " 28" AND fnlwgt = " 338409"')
print(pd.read_sql_query("SELECT * FROM sqladb", connection).head(5))
```

4.Write two filter queries

```python
cursor.execute('SELECT DISTINCT * FROM sqladb WHERE education = " Bachelors" AND age < 25 AND relationship = " Unmarried" AND sex = " Female" and native_country=" United-States"')
output = cursor.fetchall()
print(output)
```

```python
cursor.execute('SELECT DISTINCT * FROM sqladb WHERE education = " Bachelors" and age < 25 and income <= "50k" and native_country=" India"')
output = cursor.fetchall()
print(output)
```
